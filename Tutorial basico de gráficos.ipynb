### Tutorial: Habilidades Básicas de Plotagem em Python Usando Matplotlib

**Introdução:**
Matplotlib é uma biblioteca Python poderosa e popular para visualização de dados. Ela é amplamente utilizada para criar gráficos estáticos, interativos e animados. Este tutorial irá guiá-lo através do básico da plotagem com Matplotlib, desde gráficos de linhas simples até visualizações mais avançadas.

### 1. Instalação
Primeiro, certifique-se de que o Matplotlib está instalado no seu ambiente Python. Você pode instalá-lo usando pip:

```bash

```

### 2. Importando o Matplotlib
Para começar, você precisará importar o módulo `matplotlib.pyplot`, que fornece uma interface semelhante ao MATLAB para plotagem.

```python
import matplotlib.pyplot as plt
```

### 3. Gráfico de Linha Básico
Vamos começar com um gráfico de linha simples. Suponha que você tenha alguns pontos de dados que deseja plotar:

````python
import matplotlib.pyplot as plt

# Dados de exemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criar um gráfico de linha
plt.plot(x, y)

# Adicionar rótulos e título
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Linha Simples')

# Mostrar o gráfico
plt.show()
````

### 4. Personalizando Gráficos
Você pode personalizar seus gráficos adicionando marcadores, alterando estilos de linha e adicionando cores:

````python
# Personalizar o gráfico
plt.plot(x, y, color='green', marker='o', linestyle='--')

# Adicionar rótulos e título
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Linha Personalizado')

# Mostrar o gráfico
plt.show()
````

### 5. Adicionando Múltiplas Linhas
Você pode plotar várias linhas no mesmo gráfico chamando a função `plot` várias vezes:

````python

# Dados de exemplo
y2 = [1, 4, 9, 16, 25]

# Plotar duas linhas
plt.plot(x, y, label='y = [2, 3, 5, 7, 11]')
plt.plot(x, y2, label='y2 = [1, 4, 9, 16, 25]')

# Adicionar rótulos, título e legenda
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Múltiplas Linhas')
plt.legend()

# Mostrar o gráfico
plt.show()
````

### 6. Gráficos de Dispersão (Scatter Plots)
Os gráficos de dispersão são úteis para exibir pontos de dados individuais:

````python

# Criar um gráfico de dispersão
plt.scatter(x, y, color='red')

# Adicionar rótulos e título
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Dispersão Simples')

# Mostrar o gráfico
plt.show()
````

### 7. Gráficos de Barra
Os gráficos de barra são usados para exibir dados categóricos:

````python

# Dados de exemplo
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [5, 7, 3, 8, 6]

# Criar um gráfico de barra
plt.bar(categorias, valores, color='blue')

# Adicionar rótulos e título
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Gráfico de Barra Simples')

# Mostrar o gráfico
plt.show()
````

### 8. Histogramas
Os histogramas são usados para exibir a distribuição de um conjunto de dados:

````python

import numpy as np

# Dados de exemplo
dados = np.random.randn(1000)

# Criar um histograma
plt.hist(dados, bins=30, color='purple')

# Adicionar rótulos e título
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.title('Histograma')

# Mostrar o gráfico
plt.show()
````

### 9. Gráficos de Pizza
Os gráficos de pizza são úteis para mostrar proporções:

````python
# Dados de exemplo
tamanhos = [30, 20, 25, 25]
rótulos = ['Grupo A', 'Grupo B', 'Grupo C', 'Grupo D']

# Criar um gráfico de pizza
plt.pie(tamanhos, labels=rótulos, autopct='%1.1f%%', colors=['red', 'green', 'blue', 'yellow'])

# Adicionar título
plt.title('Gráfico de Pizza Simples')

# Mostrar o gráfico
plt.show()
````

### 10. Subplots
Você pode criar múltiplos gráficos em uma única figura usando `subplots`:

````python

# Criar subplots
fig, axs = plt.subplots(2, 2)

# Primeiro gráfico
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Gráfico de Linha')

# Segundo gráfico
axs[0, 1].scatter(x, y)
axs[0, 1].set_title('Gráfico de Dispersão')

# Terceiro gráfico
axs[1, 0].bar(categorias, valores)
axs[1, 0].set_title('Gráfico de Barra')

# Quarto gráfico
axs[1, 1].hist(dados, bins=30)
axs[1, 1].set_title('Histograma')

# Ajustar espaçamento
plt.tight_layout()

# Mostrar os gráficos
plt.show()
````

### 11. Salvando Gráficos
Você pode salvar seus gráficos em um arquivo usando `savefig`:

````python

# Salvar o gráfico como um arquivo PNG
plt.plot(x, y)
plt.title('Gráfico Salvo')
plt.savefig('meu_grafico.png')

# Mostrar o gráfico
plt.show()
````

Para controlar o tamanho de um gráfico salvo em formato PDF usando Matplotlib, você pode ajustar o tamanho da figura com o parâmetro `figsize` no `plt.figure()` ou `plt.subplots()` e, em seguida, salvar o gráfico usando o método `savefig()`.

Aqui está um exemplo que demonstra como fazer isso:

### Exemplo: Controlando o Tamanho do Gráfico Salvo em PDF

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerar dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Definir o tamanho da figura (largura, altura) em polegadas
fig, ax = plt.subplots(figsize=(8, 6))  # 8 polegadas de largura, 6 polegadas de altura

# Plotar os dados
ax.plot(x, y, label='Seno de x')

# Adicionar título e rótulos
ax.set_title('Exemplo de Gráfico com Controle de Tamanho')
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')

# Adicionar legenda
ax.legend()

# Salvar o gráfico em um arquivo PDF
plt.savefig('grafico_exemplo.pdf', format='pdf')

# Mostrar o gráfico (opcional)
plt.show()
```

### Explicação:

1. **`figsize=(8, 6)`**: Esse parâmetro em `plt.subplots()` define o tamanho da figura em polegadas. No exemplo, a largura é de 8 polegadas e a altura é de 6 polegadas.
2. **`plt.savefig('grafico_exemplo.pdf', format='pdf')`**: Este comando salva o gráfico no formato PDF. O tamanho do gráfico no PDF será conforme especificado em `figsize`.
3. **Qualidade e Tamanho**: Definir o tamanho da figura dessa maneira garante que o gráfico tenha a qualidade e as proporções desejadas quando salvo em PDF.

### Resultado

O gráfico será salvo como um arquivo `grafico_exemplo.pdf` com o tamanho especificado. Quando você abrir o PDF, verá que o gráfico tem 8 por 6 polegadas.

### Dicas Adicionais:

- **DPI (Pontos por Polegada)**: Se você quiser controlar ainda mais a qualidade do gráfico, pode adicionar o parâmetro `dpi` ao `savefig` para definir a resolução do PDF.
  
  ```python
  plt.savefig('grafico_exemplo.pdf', format='pdf', dpi=300)
  ```

  Isso define a resolução do gráfico em 300 DPI, que é uma qualidade de impressão padrão.

````python
import matplotlib.pyplot as plt
import numpy as np

# Gerar dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Definir o tamanho da figura (largura, altura) em polegadas
fig, ax = plt.subplots(figsize=(8, 6))  # 8 polegadas de largura, 6 polegadas de altura

# Plotar os dados
ax.plot(x, y, label='Seno de x')

# Adicionar título e rótulos
ax.set_title('Exemplo de Gráfico com Controle de Tamanho')
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')

# Adicionar legenda
ax.legend()

# Salvar o gráfico em um arquivo PDF
plt.savefig('grafico_exemplo.pdf', format='pdf')

# Mostrar o gráfico (opcional)
plt.show()
````

Para configurar o tamanho de todas as fontes em gráficos Matplotlib de forma global, você pode ajustar o parâmetro `font.size` no `rcParams` do Matplotlib. Isso permitirá que você defina um tamanho de fonte padrão para todos os elementos do gráfico, incluindo títulos, rótulos dos eixos, legendas, e rótulos de marcação dos eixos (ticks).

Aqui está um exemplo de como configurar o tamanho de todas as fontes no Matplotlib:

### Exemplo: Configurando o Tamanho Global das Fontes

```python
import matplotlib.pyplot as plt
import matplolib as mpl
import numpy as np

# Configurar o tamanho global das fontes
mpl.rcParams['font.size'] = 10  # Define o tamanho da fonte para 14
mpl.rcParams['lines.linewidth'] = 2 # espessura das linhas
# Gerar dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Criar uma figura
fig, ax = plt.subplots(figsize=(8, 6))  # 8 polegadas de largura, 6 polegadas de altura

# Plotar os dados
ax.plot(x, y, label='Seno de x')

# Adicionar título e rótulos dos eixos (os tamanhos das fontes são controlados globalmente)
ax.set_title('Exemplo de Gráfico com Fonte Global')
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')

# Adicionar legenda
ax.legend()

# Salvar o gráfico em um arquivo PDF
plt.savefig('grafico_exemplo.pdf',bbox_inches="tight")

# Mostrar o gráfico
plt.show()
```

### Explicação:

1. **`plt.rcParams['font.size'] = 14`**: Esta linha configura o tamanho global da fonte para todos os elementos do gráfico. Neste caso, o tamanho da fonte é definido como 14.

2. **Elementos Afetados**:
   - **Título (`ax.set_title`)**: O tamanho da fonte é definido como 16 explicitamente, sobrescrevendo o tamanho global, mas outros elementos usarão o tamanho global de 14.
   - **Rótulos dos Eixos (`ax.set_xlabel`, `ax.set_ylabel`)**: Usarão o tamanho da fonte global configurado.
   - **Rótulos de Marcação dos Eixos (Ticks)**: Os números ao longo dos eixos usarão o tamanho da fonte global configurado.
   - **Legenda (`ax.legend`)**: Usará o tamanho da fonte global configurado.

3. **Personalização Adicional**:
   Se você quiser configurar tamanhos de fonte diferentes para elementos específicos (como títulos ou rótulos dos eixos), pode usar parâmetros adicionais no `rcParams`, como:

   ```python
   plt.rcParams['axes.titlesize'] = 16   # Tamanho do título
   plt.rcParams['axes.labelsize'] = 14   # Tamanho dos rótulos dos eixos
   plt.rcParams['xtick.labelsize'] = 12  # Tamanho dos rótulos dos ticks do eixo X
   plt.rcParams['ytick.labelsize'] = 12  # Tamanho dos rótulos dos ticks do eixo Y
   plt.rcParams['legend.fontsize'] = 13  # Tamanho da fonte da legenda
   ```
   
   Veja mais exemplos https://matplotlib.org/stable/users/explain/customizing.html

4. **bbox_inches**

    Note que foi utilizado o comando `bbox_inches="tight"` para salvar a figura, isto é importante para evitar que o arquivo salvo não corte as fontes

### Resultado

Com esta configuração, todos os gráficos que você gerar em sua sessão de Matplotlib usarão o tamanho de fonte padrão que você especificou. Isso é útil para manter a consistência visual em relatórios e apresentações.

````python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

# Configurar o tamanho global das fontes
mpl.rcParams['font.size'] = 10  # Define o tamanho da fonte para 14
mpl.rcParams['lines.linewidth'] = 2 # espessura das linhas

# Gerar dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Criar uma figura
#função para converter polegadas em cm
def cm2inch(x,y):
    return (x/2.54,y/2.54)
    
fig, ax = plt.subplots(figsize=cm2inch(8, 6))  # 8 cm de largura, 6 cm de altura
# Plotar os dados
ax.plot(x, y, label='Seno de x')

# Adicionar título e rótulos dos eixos (os tamanhos das fontes são controlados globalmente)
ax.set_title('Exemplo de Gráfico com Fonte Global')
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')

# Adicionar legenda
ax.legend()
# Salvar o gráfico em um arquivo PDF
plt.savefig('grafico_exemplo_pdf.pdf', bbox_inches="tight")
plt.savefig('grafico_exemplo_png.png', bbox_inches="tight")
# Mostrar o gráfico
plt.show()
````

## Exemplos adicionais

Exemplo de como adicionar nomes aos eixos usando LaTeX no Matplotlib:

````python
import matplotlib.pyplot as plt

# Dados de exemplo
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

# Variáveis para legendas, rótulos e título
titulo = "Comparação de Funções"
rotulo_x = r"$\text{Eixo x:}\alpha = \int{x}dx$"
rotulo_y = r" Velocidade (m/s$^2$)"
legenda_y1 = r"$y_1 = x^2$"
legenda_y2 = r"$y_2 = x$"
cor_y1 = 'blue'
cor_y2 = 'green'

# Criar o gráfico com as variáveis
plt.plot(x, y1, label=legenda_y1, color=cor_y1, marker='o')
plt.plot(x, y2, label=legenda_y2, color=cor_y2, linestyle='--')

# Adicionar rótulos e título com LaTeX
plt.xlabel(rotulo_x)
plt.ylabel(rotulo_y)
plt.title(titulo)

# Adicionar legenda
plt.legend()

# Mostrar o gráfico
plt.show()
````
